```r
numericMatrix <- function(data) {
    vn <- names(data)
    outdata <- data.frame(row1=1:nrow(data))
    for(v in vn[2:85]) {
         if ((is.numeric(data[[v]])) & (max(data[[v]],na.rm=TRUE)>2)  ) {  
                mv <- data[[v]] 
                outdata[v] <- mv }
            
        else {
            mv <- rep(0,nrow(data))
        }
    }
    outdata$row1 <- NULL
    return(outdata)
}

```

```r
num.movies <- numericMatrix(movies_ff_v)
```

```r
options(repr.plot.width =10, repr.plot.height = 48)
par(mfrow=c(5,2))
par(cex.lab=2)
par(cex.axis=2)
for (v in names(num.movies)){
    boxplot(num.movies[v],main=v,cex=1)
}
par(mfrow=c(10,10))
```

```r
missingMatrix <- function(data) {
    vn <- names(data[2:85])
    missdata <- data.frame(row1=1:nrow(data))
    for(v in vn) {
        mv <- ifelse(is.na(data[[v]]),1,0)
        missdata[v] <- mv
    }
    missdata$row1 <- NULL
    return(missdata)
}
```

```r
movies.miss <- missingMatrix(movies_ff_v)
```
