```r
nf<- c()
for (v in names(df)){
    if ((!v %in% c('Year','month','store_nbr','item_nbr'))&(is.numeric(df[[v]])==T)&
        (max(df[[v]],na.rm=T)>1 & length(unique(df[[v]]))>3))  {
        nf<- c(nf,v)
    } 
}
```
