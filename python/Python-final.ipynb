## Q5. For which courses the percentage of male/female students is over 70%?

```python
df1['Male']=0
df1.loc[(df1['Gender']=='M'),'Male']=1
df1['Female']=0
df1.loc[(df1['Gender']=='F'),'Female']=1
df1.head()
```

```python
dfgenM=df1.groupby('CourseName').agg({'Male':sum})
dfgenM.head()
```

```python
dfgenW=df1.groupby('CourseName').agg({'Female':sum})
dfgenW.head()
```

```python
dfgen=dfgenM.join(dfgenW)
dfgen.head()
```

```python
dfgen['Total']=dfgen['Male']+dfgen['Female']
dfgen.head()
```

```python
dfgen['MalePercent']=dfgen['Male']*100/dfgen['Total']
dfgen['FemalePercent']=dfgen['Female']*100/dfgen['Total']
dfgen.head()
```

```python
dfgen[dfgen.FemalePercent>70]
```

## Q6. For each department, how many students passed with a grades over 80?

```python
dfavgdep=df1.groupby(['DepartmentName','StudentId']).agg({'degree':'mean'})
dfavgdep.head()
```

```python
dfavgdep['Avg>80']=1
dfavgdep.loc[(dfavgdep['degree']<=80),'Avg>80']=0
dfavgdep['Avg<=80']=1
dfavgdep.loc[(dfavgdep['degree']>80),'Avg<=80']=0
dfavgdep.head()
```

```python
dfavghigh=dfavgdep.groupby(['DepartmentName']).agg({'Avg>80':sum})
dfavghigh
```

```python
dfavgmed=dfavgdep.groupby(['DepartmentName']).agg({'Avg<=80':sum})
dfavgmed
```

```python
dfavgsum=pd.merge(dfavgmed,dfavghigh,on='DepartmentName')
dfavgsum
```

```python
dfavgsum['TotalStudent']=dfavgsum['Avg<=80']+dfavgsum['Avg>80']
dfavgsum['Avg<=80%']=dfavgsum['Avg<=80']*100/dfavgsum['TotalStudent']
dfavgsum['Avg>80%']=dfavgsum['Avg>80']*100/dfavgsum['TotalStudent']
dfavgsum
```

## Q7. For each department, how many students passed with a grades under 60?

```python
dfavgpass=dfavgdep.groupby(['DepartmentName']).agg({'Avg>=60':sum})
dfavgpass
```

```python
dfavgfail=dfavgdep.groupby(['DepartmentName']).agg({'Avg<60':sum})
dfavgfail
```

```python
dfavg60sum=pd.merge(dfavgpass,dfavgfail,on='DepartmentName')
dfavg60sum
```

```python
dfavg60sum['TotalStudent']=dfavg60sum['Avg<60']+dfavg60sum['Avg>=60']
dfavg60sum['Avg<60%']=dfavg60sum['Avg<60']*100/dfavg60sum['TotalStudent']
dfavg60sum['Avg>=60%']=dfavg60sum['Avg>=60']*100/dfavg60sum['TotalStudent']
dfavg60sum
```
