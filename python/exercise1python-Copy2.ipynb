```python
dfavgdep=df1.groupby(['DepartmentName','StudentId']).agg({'degree':'mean'})
dfavgdep.head()
```

```python
dfavghigh=dfavgdep.groupby(['DepartmentName']).agg({'Avg>80':sum})
dfavghigh
```

```python
dfavgmed=dfavgdep.groupby(['DepartmentName']).agg({'Avg<=80':sum})
dfavgmed
```

```python
dfavgsum=pd.merge(dfavgmed,dfavghigh,on='DepartmentName')
dfavgsum
```

```python
dfavgsum['TotalStudent']=dfavgsum['Avg<=80']+dfavgsum['Avg>80']
dfavgsum['Avg<=80%']=dfavgsum['Avg<=80']*100/dfavgsum['TotalStudent']
dfavgsum['Avg>80%']=dfavgsum['Avg>80']*100/dfavgsum['TotalStudent']
dfavgsum
## end of question 2 f
```
